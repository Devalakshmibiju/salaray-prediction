```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import math
```

```python
# Data (Years of Experience and Salary)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable (Years of Experience)
y = np.array([40000, 45000, 50000, 55000, 60000])  # Dependent variable (Salary)
```

```python

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Make a prediction for 6 years of experience
experience = np.array([[6]])
predicted_salary = model.predict(experience)
print(f"Predicted Salary for 6 years of experience: ${predicted_salary[0]:.2f}")
```

```python

# Predict salaries for the given X values
y_pred = model.predict(X)
```

```python

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
```

```python
# Calculate Root Mean Squared Error (RMSE)
rmse = math.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
# Calculate R-squared (R²)
r2 = r2_score(y, y_pred)
print(f"R-squared (R²): {r2:.2f}")
```

```python
# Plot the regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (in $)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
